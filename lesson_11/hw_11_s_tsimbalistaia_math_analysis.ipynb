#### 1. Исследовать ряд на сходимость, используя признак д’Аламбера:
$\sum\limits_{n=1}^\infty\frac{n^n}{(n!)^2}$

$\lim_{n \to \infty}\frac{\frac{(n+1)^{n+1}}{((n+1)!)^2}}{\frac{n^n}{(n!)^2}}$ = $\lim_{n \to \infty}\frac{(n+1)^{n+1}(n!)^{2}}{((n+1)!)^2n^n}$ =  $\lim_{n \to \infty}\frac{(n+1)^n(n+1)n!n!}{n!(n+1)n!(n+1)n^n}$ =  $\lim_{n \to \infty}\frac{(n+1)^n}{(n+1)n^n}$ = $\lim_{n \to \infty}\frac{1}{n+1}\lim_{n \to \infty}(\frac{n+1}{n})^n$ = $\lim_{n \to \infty}\frac{1}{n+1}\lim_{n \to \infty}(1 + \frac{1}{n})^n$ = $0e$ = $0; 0<1$ $\Rightarrow$ ряд сходитcя

#### 3. Исследовать ряд на сходимость, используя признак Лейбница:
$\sum\limits_{n=1}^\infty\frac{(-1)^n}{n + ln(n)}$

$\lim_{n\to\infty}\frac{(-1)^n}{n + ln(n)}$ = $0$ $\Rightarrow$ ряд сходитcя

Так как имеем знакочередующийся ряд, нам надо теперь выяснить, он сходится условно или абсолютно. Для этого используем радикальный признак Коши.

$\lim_{n\to\infty}\sqrt[n]{\frac{1}{n + ln (n)}} = \lim_{n\to\infty}\frac{1}{\sqrt[n]{n + ln (n)}} < 1$

Получается модуль исходного ряда сходится по радикальному признаку Коши.

Значит ряд $\sum\limits_{n=1}^\infty\frac{(-1)^n}{n + ln(n)}$ сходится абсолютно.
