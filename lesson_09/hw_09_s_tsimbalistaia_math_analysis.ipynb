1. **Исследовать функцию на условный экстремум**

$U = 3 - 8x + 6y,  \; если  \; x^2 + y^2 = 36$

Решение

$U = 3 - 8x + 6y,  \; если  \; x^2 + y^2 = 36$

$L( \lambda_1, x, y)  = 3 - 8x + 6y + \lambda_1 * (x^2 + y^2 - 36)$

$\left\{
  \begin{array}{ccc}
    L'_x = -8 + 2\lambda_1x& = & 0 \\
    L'_y = 6 + 2\lambda_1y & = & 0 \\
    L'_{\lambda_1} =x^2 + y^2 - 36 & = & 0 \\
  \end{array}
\right.$

$\left\{
  \begin{array}{ccc}
   x & = & \frac{4}{\lambda_1} \\
   y & = & \frac{-3}{\lambda_1} \\
   (\frac{4}{\lambda_1})^2 + (\frac{-3}{\lambda_1})^2 - 36 & = & 0 \\
  \end{array}
\right.$

$\left\{
  \begin{array}{ccc}
   x & = & \frac{4}{\lambda_1} \\
   y & = & \frac{-3}{\lambda_1} \\
   \lambda_1 & = & \pm\frac{5}{6} \\
  \end{array}
\right.$

$(\frac{5}{6}; \frac{24}{5}; \frac{-18}{5});  \; (\frac{-5}{6}; \frac{-24}{5}; \frac{18}{5});$

$\left\{
  \begin{array}{ccc}
    L'_x & = & -8 + 2\lambda_1x \\
    L'_y & = & 6 + 2\lambda_1y  \\
    L'_y & = & x^2 + y^2 - 36 & \\
  \end{array}
\right.$

$ L''_{xx} = 2\lambda_1 \; L''_{yy} = 2\lambda_1 \; L''_{\lambda_1 \lambda_1} =0$

$ L''_{xy} = 0 \; L''_{x\lambda_1} = 2x \; L''_{y \lambda_1} =2y$

$\left|
  \begin{array}{ccc}
     0 & 2x & 2y \\
     2x & 2\lambda_1 & 0 \\
     2y & 0 & 2\lambda_1 \\
  \end{array}
\right| = 0 *\left|
  \begin{array}{ccc}
     2\lambda_1 &  & 0\\
     0 &  & 2\lambda_1 \\
  \end{array}
\right| - 2x* \left|
  \begin{array}{ccc}
     2x &  & 0\\
     2y &  & 2\lambda_1 \\
  \end{array}
\right| + 2y * \left|
  \begin{array}{ccc}
     2x &  & 2\lambda_1\\
     2y &  & 0 \\
  \end{array}
\right| = -8\lambda_1(x^2 + y^2)$

$(\frac{5}{6}; \frac{24}{5}; \frac{-18}{5});  \; (\frac{-5}{6}; \frac{-24}{5}; \frac{18}{5});$

$-8\lambda_1(x^2 + y^2) = -8 * (\frac{5}{6}) * ((\frac{24}{5})^2 + (\frac{-18}{5})^2)= -240 <0 - минимум$

$-8\lambda_1(x^2 + y^2) = -8 * (\frac{-5}{6}) * ((\frac{-24}{5})^2 + (\frac{18}{5})^2)= 240 > 0 - максимум$

$Ответ: (\frac{5}{6}; \frac{24}{5}; \frac{-18}{5}) - минимум;  \; (\frac{-5}{6}; \frac{-24}{5}; \frac{18}{5}) -максимум$


2. **Исследовать функцию на условный экстремум**

$U = 2x^2 + 12xy +32y^2 +15,  \; если  \; x^2 + 16y^2 = 64$

Решение

$U = 2x^2 + 12xy +32y^2 +15,  \; если  \; x^2 + 16y^2 = 64$

Упростим исходное выражение с помощью условия:

$U = 2x^2 + 12xy +32y^2 +15 = 2 (x^2 + 16y^2) + 12xy +15 = 2 * 64 + 12xy +15 = 128 + 12xy + 15 = 143 + 12xy$

$L( \lambda_1, x, y)  = 143 + 12xy + \lambda_1 * (x^2 + 16y^2 - 64)$

$\left\{
  \begin{array}{ccc}
    L'_x = 12y + 2\lambda_1x& = & 0 \\
    L'_y = 12x + 16\lambda_1 * 2y & = & 0 \\
    L'_{\lambda_1} =x^2 + 16y^2 - 64 & = & 0 \\
  \end{array}
\right.$

$\left\{
  \begin{array}{ccc}
   6y + \lambda_1x & = & 0 \\
   3x + 8\lambda_1y & = & 0 \\
   x^2 + 16y^2 - 64& = & 0 \\
  \end{array}
\right.$

$\left\{
  \begin{array}{ccc}
   \lambda_1 & = & \frac{-6y}{x} \\
   \lambda_1 & = & \frac{-3x}{8y} \\
   x^2 + 16y^2 - 64& = & 0 \\
  \end{array}
\right.$

Приравняем друг другу первые два уравнения и выразим x через y:

$\frac{-6y}{x} = \frac{-3x}{8y}$

$48y^2 = 3x^2$

$x^2 = 16y^2$

$x = \pm4y$

Подставим найденный x в третье уравнение:

$(\pm4y)^2 + 16y^2 - 64 = 0$

$ 16y^2 + 16y^2 - 64 = 0$

$ 32y^2 = 64$

$ y^2 = 2$

$ y = \pm \sqrt{2}$

Подставим значения и найдем стационарные точки:

$(\frac{-3}{2}; 4\sqrt{2}; \sqrt{2});  \; (\frac{3}{2}; -4\sqrt{2}; \sqrt{2}); \;(\frac{3}{2}; 4\sqrt{2}; -\sqrt{2}); 
\;(\frac{-3}{2}; -4\sqrt{2}; -\sqrt{2});$

Найдем 2-е производные:


$ L''_{xx} = \lambda_1\; L''_{yy} = 8\lambda_1 \; L''_{\lambda_1 \lambda_1} =0$

$ L''_{xy} = 6 \; L''_{x\lambda_1} = x \; L''_{y \lambda_1} =8y$

$\left|
  \begin{array}{ccc}
     0 & x & 8y \\
     x & \lambda_1 & 6 \\
     8y & 6 & 8\lambda_1 \\
  \end{array}
\right| = 0 *\left|
  \begin{array}{ccc}
     \lambda_1 &  & 6\\
     6&  & 8\lambda_1 \\
  \end{array}
\right| - x* \left|
  \begin{array}{ccc}
     x &  & 6\\
     8y &  & 8\lambda_1 \\
  \end{array}
\right| + 8y * \left|
  \begin{array}{ccc}
     x &  & \lambda_1\\
     8y &  & 6 \\
  \end{array}
\right| = -8x^2\lambda_1 + 96xy -64\lambda_1y^2$

Подставим стационарные точки:

$(\frac{-3}{2}; 4\sqrt{2}; \sqrt{2});  \; (\frac{3}{2}; -4\sqrt{2}; \sqrt{2}); \;(\frac{3}{2}; 4\sqrt{2}; -\sqrt{2}); 
\;(\frac{-3}{2}; -4\sqrt{2}; -\sqrt{2});$

$-8x^2\lambda_1 + 96xy -64\lambda_1y^2 = -8( 4\sqrt{2})^2*(\frac{-3}{2}) + 96 * (4\sqrt{2}) * (\sqrt{2}) -64(\frac{-3}{2})*(\sqrt{2})^2 = 1152 > 0 - максимум$

$-8x^2\lambda_1 + 96xy -64\lambda_1y^2 = -8( -4\sqrt{2})^2*(\frac{3}{2}) + 96 * (-4\sqrt{2}) * (\sqrt{2}) -64(\frac{3}{2})*(\sqrt{2})^2 = -1152 < 0 - минимум$

$-8x^2\lambda_1 + 96xy -64\lambda_1y^2 = -8( 4\sqrt{2})^2*(\frac{3}{2}) + 96 * (4\sqrt{2}) * (-\sqrt{2}) -64(\frac{3}{2})*(-\sqrt{2})^2 = -1152 < 0 - минимум$

$-8x^2\lambda_1 + 96xy -64\lambda_1y^2 = -8( -4\sqrt{2})^2*(\frac{-3}{2}) + 96 * (-4\sqrt{2}) * (-\sqrt{2}) -64(\frac{-3}{2})*(-\sqrt{2})^2 = 1152 > 0 - максимум$

$Ответ: (\frac{-3}{2}; 4\sqrt{2}; \sqrt{2}); - максимум;  \; (\frac{3}{2}; -4\sqrt{2}; \sqrt{2}) -минимум; \;
(\frac{3}{2}; 4\sqrt{2}; -\sqrt{2}) - минимум;\; (\frac{-3}{2}; -4\sqrt{2}; -\sqrt{2})- максимум;$

3. **Найти производную функции $U =x^2 + y^2 + z^2 \;  по \; направлению \; вектора  \; \overline{с}(-9;8;-12)\;  в\;  точке \; M(8;-12;9)$**

Решение

$|\overline{с}| = \sqrt{(-9)^2 + 8^2 + (-12)^2} = \sqrt{289} = 17$

$\overline{с_0} = (\frac{-9}{17};\frac{8}{17};\frac{-12}{17})$

$U =x^2 + y^2 + z^2$

$U'_{x} = 2x$

$U'_{y} = 2y$

$U'_{z} = 2z$

$gradU|_{(8;-12;9)} = (16; -24; 18)$

$U'_\overline{с}|_{(8;-12;9)} = \frac{-9 *16}{17} + \frac{-24*8}{17} + \frac{-12 * 18}{17} = \frac{-552}{17}$

$Ответ:  \frac{-552}{17}$

4. **Найти производную функции $U = e^{x^2 + y^2 + z^2} \;  по \; направлению \; вектора  \; \overline{d}(4;-13;-16)\;  в\;  точке \; L(-16;4;-13)$**

Решение

$|\overline{d}| = \sqrt{(4)^2 + (-13)^2 + (-16)^2} = \sqrt{441} = 21$

$\overline{d_0} = (\frac{4}{21};\frac{-13}{21};\frac{-16}{21})$

$U =e^{x^2 + y^2 + z^2}$

$U'_{x} = 2x * e^{x^2 + y^2 + z^2}$

$U'_{y} = 2y * e^{x^2 + y^2 + z^2}$

$U'_{z} = 2z * e^{x^2 + y^2 + z^2}$

$Найдем \;  x^2 + y^2 + z^2 \; в\;  точке \; L(-16;4;-13):$

$(-16)^2 + 4^2 + (-13)^2  = 441$

$gradU|_{(-16;4;-13)} = (-32e^{441}; 8e^{441}; -26e^{441})$

$\overline{d_0} = (\frac{4}{21};\frac{-13}{21};\frac{-16}{21})$

$U'_\overline{d}|_{(-16;4;-13)} = \frac{4}{21} * (-32e^{441})  + \frac{-13}{21} * (8e^{441})+ \frac{-16}{21} * (-26e^{441})= \frac{184e^{441}}{21}$

$Ответ:  \frac{184e^{441}}{21}$
